```python
# Papagaio

# features [1 sim, 0 nao]
# tem bico?
# risco extino?
# faz cocoric?

papagaio1 = [1, 1, 0]
papagaio2 = [1, 1, 0]
papagaio3 = [1, 1, 1]
```

```python
# Galinha 

# feactures [1 sim, 0 nao]
# tem bico?
# risco extino?
# faz cocoric?

galinha1 = [1, 0, 1]
galinha2 = [1, 0, 1]
galinha3 = [1, 1, 1]
```

```python
dados = [ papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
classes = [1, 1, 1, 0, 0, 0]
```

```python
dados
```

```python
# TREINAMENTO DO MODELO

! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

PREVER UM ANIMAL

```python
# features [1 sim, 0 nao]
# tem bico?
# risco extino
# faz cocoric?

# tem bico, esta em extino e no faz cocoric

animal_misterioso = [1, 1, 0]
previsao = modelo.predict([animal_misterioso])
```

# Calculadora a Acurcia do Modelo

# Dados e Rtulos

```python
# dados de treino x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1,galinha2, galinha3]
# rotulos (classes) de treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste 

misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]


```

```python
# previses -  dados de testes
teste_x = [misterio1, misterio2, misterio3] 
# prevendo os testes 
previsoes = modelo.predict(teste_x)
```

```python
# rotulos teste (classes) de teste (avaliao - aqui  a sua provinha)
teste_y = [0, 1, 1]

```

```python
# importando modulo que j traz a funo acuracia encapsulada
from sklearn.metrics import accuracy_score


# definindo uma variavel para receber o retorno da funo acuracia - 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100

# mostrando a curcia 
print(f" Acuracia: {taxa_de_acerto:.2f}%")
```
