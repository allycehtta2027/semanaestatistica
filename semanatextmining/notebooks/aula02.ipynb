```python
Table: funcionario
CREATE TABLE funcionario (
    id INT PRIMARY KEY AUTO_INCREMENT,
    nome VARCHAR(50)
);

 Table: cliente
CREATE TABLE cliente (
    id INT PRIMARY KEY AUTO_INCREMENT,
    nome VARCHAR(50),
    telefone CHAR(11),
    email VARCHAR(50),
    senha VARCHAR(200)
);

Table: servico
CREATE TABLE servico (
    id INT PRIMARY KEY AUTO_INCREMENT,
    nome VARCHAR(50),
    valor DECIMAL(9,3)
);

 Table: agendamento
CREATE TABLE agendamento (
    id INT PRIMARY KEY AUTO_INCREMENT,
    data DATE,
    horario TIME,
    fk_cliente_id INT,
    fk_servico_id INT,
    fk_funcionario_id INT,
    FOREIGN KEY (fk_cliente_id) REFERENCES cliente(id),
    FOREIGN KEY (fk_servico_id) REFERENCES servico(id),
    FOREIGN KEY (fk_funcionario_id) REFERENCES funcionario(id)
);
```

```python
SELECT 
a.data,
a.horario,
c.nome AS cliente_nome,
s.nome AS servico_nome,
.nome AS funcionario_nome
FROM agendamento a
 cliente c ON a.fk_cliente_id = c.id
JOIN servico s ON a.fk_servico_id = s.id
 funcionario f ON a.fk_funcionario_id = f.id
```

QUAL A DIFERENÇA ENTRE FLASK MAIUSCULO E FLASK MINUSCULO
Um aplicativo web (ou outro qualquer) deve ser escrito de forma clara, para facilitar sua expansão e manutenção. Uma das formas usadas pelo Flask para implementar esse estratégia é a de colocar código python e html separados. Os templates, como vimos, são modelos ou estruturas básicas que podem ser preenchidas dinamicamente, de acordo com as requisições. Esse é o chamado de modelo de separação entre lógica de negócio e lógica de exibição (business and presentation logic). Templates são tratados por um dos módulos que compõem o Flask: o módulo Jinja2.

# suponha que temos a variável titulo = "a casa da tia maria"
# essa string pode ser exibida dentro de uma tag <h1>
<h1> {{ titulo }} </h1>
↳ a casa da tia maria

# para maísculas na primeira letra de cada palavra
<h1> {{ titulo | title() }} </h1>
↳ A Casa da tia maria

```python
from flask import Flask

app = Flask(__name__)

# Home route
@app.route('/')
def index():
    return "Estilo do Rei Barbearia - Página Principal"

# Route for adding a new employee
@app.route('/novofuncionario')
def novo_funcionario():
    return "Página para adicionar um novo funcionário"

# Route for adding a new client
@app.route('/novocliente')
def novo_cliente():
    return "Página para adicionar um novo cliente"

# Route for adding a new service
@app.route('/novoservico')
def novo_servico():
    return "Página para adicionar um novo serviço"

# Route for creating a new appointment
@app.route('/novoagendamento')
def novo_agendamento():
    return "Página para adicionar um novo agendamento"

# Route for login
@app.route('/login')
def login():
    return "Página de login"

# Route for logout
@app.route('/logout')
def logout():
    return "Logout efetuado com sucesso"

if __name__ == '__main__':
    app.run(debug=True)
```
