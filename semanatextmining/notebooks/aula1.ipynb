```python
import requests

# URL da API com o CEP desejado
url = 'https://viacep.com.br/ws/72705030/json/'

# Fazendo a requisição GET para a API
response = requests.get(url)

# Verifica se a requisição foi bem-sucedida (status code 200)
if response.status_code == 200:
    # Convertendo a resposta em JSON
    data = response.json()
    
    # Exibindo os dados
    print("Dados do CEP:")
    for key, value in data.items():
        print(f"{key}: {value}")
else:
    print("Erro ao fazer a requisição:", response.status_code)
```

Diferença entre Pétala e Sepala
Sépala: Folhas modificadas, geralmente de coloração verde, localizadas abaixo das pétalas e com função de proteger o botão floral enquanto ainda não abriu; Pétala: Unidade da corola, pode possuir cores vivas e marcantes e sua função é atrair agentes polinizadores.

```python
# mostrar quantas variedades da íris
df['variety'].value_counts()
```

```python
# Aplicação do List comprehension 
# [] - List 
# string --> s
df.columns = [s.replace('.','_') for s in df.columns]
```

```python
df.head(2)
```

```python
# pesquisa
# copia independente (copy() não agetra as alterações originais
df_setosa = df.query('variety =="Setosa" ').copy()
df_virginica =df.query('variety == "Virginica"').copy()
df_versicolor = df.query('variety == "Versicolor"').copy()

print(df_setosa)
print(df_virginica)
print(df_versicolor)

# criar uma lista com nomes das colunas de medidas.
medidas = ['sepal_length','sepal_width','petal_length','petal_width']

```
